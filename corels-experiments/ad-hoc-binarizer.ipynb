# Churn data set - Ad hoc binarization for CORELS

```python
file_name = 'data/churn_prob_out_35.csv' 
ROW_LIMIT = None # 10000
expected_data_type = {
    'CHURN': 'object'
    }

def discretize( x, bins=3 ):
    classes = [ (k / (bins-1)) for k in range(bins-1) ] + [1.0]
    diffs = [ abs(x-k) for k in classes ]
    # print(diffs)
    arg_min = diffs.index(min(diffs))
    return classes[arg_min]

def convert_label(char, pos_value='T'):
    if (char == pos_value):
        return 1
    else:
        return 0

def greater_than(number, ref=0):
    if (number >= ref):
        return 1
    else:
        return 0

# def discretize_float( column, min_float, max_float, step_size ):
```

## Read data

```python
df = pd.read_csv(file_name, nrows=ROW_LIMIT)
print('Read {} rows from {}.'.format(len(df), file_name))

df.info(verbose=True)
df.head()


```

```python
# print(min(df['Est Income']))
col = 'Usage'
print(min(df[col]), max(df[col]))

col = 'Est Income'
for y in range(30000, 120000, 30000):
    df[col + ' <= ' + str(y)] = df[col].apply(lambda x: int(x <= y))
    df[col + ' >= ' + str(y)] = df[col].apply(lambda x: int(x >= y))

col = 'Age'
for y in range(30, 90, 30):
    df[col + ' <= ' + str(y)] = df[col].apply(lambda x: int(x <= y))
    df[col + ' >= ' + str(y)] = df[col].apply(lambda x: int(x >= y))

col = 'Usage'
for y in range(30, 180, 30):
    df[col + ' <= ' + str(y)] = df[col].apply(lambda x: int(x <= y))
    df[col + ' >= ' + str(y)] = df[col].apply(lambda x: int(x >= y))

col = 'Gender'
pos_val = 'F'
df[col + ' = ' + pos_val] = df[col].apply(convert_label, pos_value=pos_val)

col = 'Car Owner'
pos_val = 'Y'
df[col + ' = ' + pos_val] = df[col].apply(convert_label, pos_value=pos_val)

for col in ['Status', 'Children', 'Paymethod', 'RatePlan', '3_Class']:
    for v in df[col].unique():
            df[col + ' = ' + str(v)] = df[col].apply(lambda x: int(x == v))



# df['Est Income <= 30000'] = df['Est Income'].apply(lambda x: int(x >= 30000))

# print(df.columns)

binary_columns = ['Est Income <= 30000', 'Est Income >= 30000', 'Est Income <= 60000',
       'Est Income >= 60000', 'Est Income <= 90000', 'Est Income >= 90000',
       'Age <= 30', 'Age >= 30', 'Age <= 60', 'Age >= 60', 'Usage <= 30',
       'Usage >= 30', 'Usage <= 60', 'Usage >= 60', 'Usage <= 90',
       'Usage >= 90', 'Usage <= 120', 'Usage >= 120', 'Usage <= 150',
       'Usage >= 150', 'Gender = F', 'Status = S', 'Status = M', 'Status = D',
       'Car Owner = Y', 'Children = 1', 'Children = 2', 'Children = 0',
       'Paymethod = CC', 'Paymethod = CH', 'Paymethod = Auto', 'RatePlan = 3',
       'RatePlan = 2', 'RatePlan = 1', 'RatePlan = 4', '3_Class = 1.0']

df = df[binary_columns]
df.info()
df.head()

df.to_csv('data/churn4CORELS.csv', index=False)


```
